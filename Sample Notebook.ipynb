```python
class VortexaSDK:
    '''
    This is a mock Vortexa SDK.
    '''
    
    def __init__(self):
        self.cms = pd.read_csv('./resources/summary_cargo_movements.csv')
    
    def read_cargo_movements(self, vessel=None, destination=None, origin=None):
        all_cms = self.cms
        
        if vessel != None:
            all_cms = all_cms[all_cms['PRIMARY VESSEL'] == vessel]
            
        return all_cms
```
